```python
def verify_team_names(df, system_name, crosswalk_df):
    """
    Verify that all team names in a projection system's dataframe exist in the crosswalk
    """
    # Get unique teams from both home and away columns
    home_teams = set(df['Home Team'].unique())
    away_teams = set(df['Away Team'].unique())
    all_teams = home_teams.union(away_teams)
    
    # Get the corresponding column name from crosswalk
    crosswalk_col = system_name.lower()
    crosswalk_teams = set(crosswalk_df[crosswalk_col].unique())
    
    # Find missing teams
    missing_teams = all_teams - crosswalk_teams
    
    print(f"\n{system_name} Team Name Verification:")
    print(f"Total unique teams in {system_name}: {len(all_teams)}")
    print(f"Total teams in crosswalk: {len(crosswalk_teams)}")
    if missing_teams:
        print(f"WARNING: Missing teams in crosswalk: {missing_teams}")
        return False
    print("✓ All teams found in crosswalk")
    return True

def verify_team_level_transform(df, system_name):
    """
    Verify that dataframe has been properly transformed to team level
    """
    print(f"\n{system_name} Team Level Transform Verification:")
    
    # Check for required columns
    required_cols = [
        'Game',
        'Team',
        f'spread_{system_name.lower()}',
        f'win_prob_{system_name.lower()}',
        f'projected_total_{system_name.lower()}'
    ]
    
    missing_cols = [col for col in required_cols if col not in df.columns]
    
    if missing_cols:
        print(f"WARNING: Missing required columns: {missing_cols}")
        return False
    
    # Verify each game has exactly two teams
    game_counts = df['Game'].value_counts()
    invalid_games = game_counts[game_counts != 2]
    
    if not invalid_games.empty:
        print("WARNING: Games with incorrect number of teams:")
        print(invalid_games)
        return False
    
    print("✓ Data properly transformed to team level")
    return True

def verify_game_id_format(df, system_name):
    """
    Verify that Game IDs follow the format "Home Team vs Away Team"
    """
    print(f"\n{system_name} Game ID Format Verification:")
    
    # Create expected Game IDs
    expected_games = df.apply(lambda x: f"{x['Home Team']} vs {x['Away Team']}", axis=1)
    
    # Compare with actual Game column if it exists
    if 'Game' in df.columns:
        mismatches = df[df['Game'] != expected_games]
        if not mismatches.empty:
            print(f"WARNING: Found {len(mismatches)} games with incorrect format:")
            print(mismatches[['Game', 'Home Team', 'Away Team']].head())
            return False
    else:
        print("WARNING: No 'Game' column found")
        return False
    
    print("✓ All game IDs properly formatted")
    return True

def run_checkpoint_verification(projection_dfs, crosswalk_df):
    """
    Run all verification checks for the first checkpoint
    """
    all_passed = True
    
    for system_name, df in projection_dfs.items():
        print(f"\n{'='*50}")
        print(f"Verifying {system_name} data:")
        print('='*50)
        
        # Transform data to team level first
        transformed_df = transform_to_team_level(df, system_name.lower())
        
        # Run verifications
        names_check = verify_team_names(df, system_name, crosswalk_df)
        transform_check = verify_team_level_transform(transformed_df, system_name)
        format_check = verify_game_id_format(df, system_name)
        
        if not all([names_check, transform_check, format_check]):
            all_passed = False
            print(f"\n❌ {system_name} failed one or more verification checks")
        else:
            print(f"\n✓ {system_name} passed all verification checks")
    
    return all_passed

# Create dictionary of projection systems
projection_systems = {
    'Barttorvik': barttorvik_df,
    # 'Drating': dratings_df,
    'EvanMiya': evanmiya_df,
    'New_KenPom': kenpom_df,
    # 'Massey': massey_df
}

# Run verification
checkpoint_passed = run_checkpoint_verification(projection_systems, crosswalk_df)

print("\nFinal Checkpoint Status:")
if checkpoint_passed:
    print("✓ All systems passed verification checks")
else:
    print("❌ Some systems failed verification checks")
```

```python
import pandas as pd
import numpy as np

def custom_round(x):
    """Custom rounding function that always rounds .5 up"""
    if pd.isnull(x):
        return np.nan
    elif x > 0:
        return np.floor(x + 0.5)
    elif x < 0:
        return np.ceil(x - 0.5)
    else:
        return 0
def transform_to_team_level(df, system_name):
    if df is None or df.empty:
        return None

    # Create game identifier with consistent format
    df['Game'] = df.apply(lambda x: f"{x['Home Team']} vs {x['Away Team']}", axis=1)

    # Create records for home teams
    home_df = pd.DataFrame({
        'Game': df['Game'],
        'Team': df['Home Team'],
        f'spread_{system_name}': df['Home Team Spread'].astype(float),
        f'win_prob_{system_name}': df['Home Team Win Probability'].astype(float),
        f'projected_total_{system_name}': df['Projected Total'].astype(float)
    })

    # Create records for away teams
    away_df = pd.DataFrame({
        'Game': df['Game'],
        'Team': df['Away Team'],
        f'spread_{system_name}': df['Away Team Spread'].astype(float),
        f'win_prob_{system_name}': df['Away Team Win Probability'].astype(float),
        f'projected_total_{system_name}': df['Projected Total'].astype(float)
    })

    # Combine and return
    return pd.concat([home_df, away_df], ignore_index=True)

def process_odds_data():
    """Process odds data from all sources"""
    moneyline_df = get_moneyline_odds()
    spread_df = get_spread_odds()
    totals_df = get_totals_odds()
    
    combined_odds = pd.DataFrame()
    
    # Process moneyline odds
    if moneyline_df is not None:
        # Create consistent game identifier
        moneyline_df['Game'] = moneyline_df.apply(lambda x: f"{x['Home Team']} vs {x['Away Team']}", axis=1)
        
        # Process home teams
        home_ml = pd.DataFrame({
            'Game': moneyline_df['Game'],
            'Team': moneyline_df['Home Team'],
            'Opening Moneyline': moneyline_df['Home Team Moneyline']
        })
        
        # Process away teams
        away_ml = pd.DataFrame({
            'Game': moneyline_df['Game'],
            'Team': moneyline_df['Away Team'],
            'Opening Moneyline': moneyline_df['Away Team Moneyline']
        })
        
        combined_odds = pd.concat([home_ml, away_ml], ignore_index=True)
        
        # Calculate implied probability
        combined_odds['Opening Moneyline'] = pd.to_numeric(combined_odds['Opening Moneyline'], errors='coerce')
        combined_odds['Moneyline Win Probability'] = combined_odds['Opening Moneyline'].apply(
            lambda x: 100/(x + 100) if x > 0 else -x/(-x + 100) if not pd.isna(x) else np.nan)
    
    # Process spread odds
    if spread_df is not None:
        spread_df['Game'] = spread_df.apply(lambda x: f"{x['Home Team']} vs {x['Away Team']}", axis=1)
        
        # Process home teams
        home_spread = pd.DataFrame({
            'Game': spread_df['Game'],
            'Team': spread_df['Home Team'],
            'Opening Spread': spread_df['Home Team Spread']
        })
        
        # Process away teams
        away_spread = pd.DataFrame({
            'Game': spread_df['Game'],
            'Team': spread_df['Away Team'],
            'Opening Spread': spread_df['Away Team Spread']
        })
        
        spread_data = pd.concat([home_spread, away_spread], ignore_index=True)
        
        if not combined_odds.empty:
            combined_odds = pd.merge(combined_odds, spread_data, on=['Game', 'Team'], how='outer')
        else:
            combined_odds = spread_data
    
    # Process totals
    if totals_df is not None:
        totals_df['Game'] = totals_df.apply(lambda x: f"{x['Home Team']} vs {x['Away Team']}", axis=1)
        totals_data = totals_df[['Game', 'Projected Total']].rename(columns={'Projected Total': 'theoddsapi_total'})
        
        if not combined_odds.empty:
            combined_odds = pd.merge(combined_odds, totals_data, on='Game', how='outer')
        else:
            combined_odds = totals_data
    
    return combined_odds

def main():
    # Process projection data
    projection_df = process_projection_data()
    
    # Process odds data
    odds_df = process_odds_data()
    
    # Merge projection and odds data
    if odds_df is not None:
        final_df = pd.merge(projection_df, odds_df, on=['Game', 'Team'], how='outer')
    else:
        final_df = projection_df
    
    # Calculate average total across all systems
    total_cols = [col for col in final_df.columns if 'projected_total_' in col or col == 'theoddsapi_total']
    if total_cols:
        final_df['average_total'] = final_df[total_cols].mean(axis=1)
    
    # Calculate Moneyline Edge
    if 'Moneyline Win Probability' in final_df.columns:
        win_prob_cols = [col for col in final_df.columns if col.startswith('win_prob_')]
        final_df['median_win_prob'] = final_df[win_prob_cols].median(axis=1)
        final_df['Moneyline Edge'] = (final_df['median_win_prob'] - final_df['Moneyline Win Probability']).round(4)
    
    # Sort columns to match desired format
    desired_columns = [
        'Game', 'Team',
        'spread_barttorvik', 'win_prob_barttorvik', 'projected_total_barttorvik',
        # 'spread_drating', 'win_prob_drating', 'projected_total_drating',
        # 'spread_evanmiya', 'win_prob_evanmiya', 'projected_total_evanmiya',
        'spread_kenpom', 'win_prob_kenpom', 'projected_total_kenpom',
        # 'spread_massey', 'win_prob_massey', 'projected_total_massey',
        'Opening Moneyline', 'Moneyline Win Probability', 'Opening Spread',
        'theoddsapi_total', 'Moneyline Edge', 'average_total'
    ]
    
    final_df = final_df.reindex(columns=[col for col in desired_columns if col in final_df.columns])
    
    # Clean up the dataframe
    final_df = final_df.dropna(subset=['Game', 'Team'])
    final_df = final_df.drop_duplicates(subset=['Game', 'Team'])
    
    return final_df

if __name__ == "__main__":
    final_df = main()
    final_df.to_csv('final_df.csv', index=False)
    print("ETL process completed successfully.")
```
